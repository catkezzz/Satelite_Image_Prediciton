## Import Library

```python
import pickle
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

## Load Model

```python
with open('best_cnn_model.pkl', 'rb') as file_1:
  model = pickle.load(file_1)
```

## Load Gambar dan Prediksi

Buat fungsi untuk prediksi. Data inferens adalah gambar satelit untuk situasi cloudy. Sumber: https://blog.mapbox.com/cloudy-satellite-imagery-a-chronicle-of-imagery-collection-over-nepal-post-disaster-8830e3cbd1f

```python
# Create a function for test a new image

def prediction(file):
  img = tf.keras.utils.load_img(file, target_size=(64,64))
  x = tf.keras.utils.img_to_array(img)/255
  
  plt.imshow(img)

  x = np.expand_dims(x, axis=0)

  images = np.vstack([x])
  y_pred_proba = model.predict(images)
  y_pred_class = np.argmax(y_pred_proba[0])
  print(y_pred_class)
  class_name = list(['Cloudy', 'Desert', 'Green area', 'Water'])
  y_pred_class_name = class_name[y_pred_class]

  print('Prediction : {}'.format(y_pred_class_name))
```

```python
prediction('img_inferens.jpg')
```
